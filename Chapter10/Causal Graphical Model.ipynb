```python
from causalgraphicalmodels import CausalGraphicalModel
```

```python
Model = CausalGraphicalModel(
    nodes=["Job", "Smartwork", "Hardwork", "Reward", "Promotion"],
    edges=[
        ("Job", "Smartwork"), 
        ("Job", "Hardwork"), 
        ("Smartwork", "Reward"),
        ("Hardwork", "Reward"), 
        ("Reward", "Promotion")
    ]
)
Model.draw()
```

```python
# get the distribution 
print(Model.get_distribution())
```

```python
# get all the conditional independence relationships 
Model.get_all_independence_relationships()
```

```python

# Let's intervene it by "Reward"
Intervene = Model.do("Reward")
Intervene.draw()
```
